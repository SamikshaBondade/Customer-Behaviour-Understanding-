```python
# check the heading names and keep always heading names as lower case and don't use space insted use underscore 

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
# create a new column age_group 

labels = ['Young_Adult', 'Adult', 'Middle_Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# purchase frequency days - new columns 

frequency_maping = {
    'Fornightly': 14,
    'Weekly': 7, 
    'Monthly': 30, 
    'Quarterly': 90, 
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90 
}
df['purchase_frequency_days'] = df['frequency_of_purchase'].map(frequency_maping)
```
